```python
from nas_201_api import NASBench201API as API
from nas_201_api import ResultsCount
from tqdm import tqdm
import statistics as stat
```

```python
#api = API('../data/NAS-Bench-201-v1_0-e61699.pth')
api = API('../data/NAS-Bench-201-v1_1-096897.pth')
```

```python
accuracy = {}
for i in tqdm(range(len(api))):
    results = api.query_by_index(i, 'cifar10', hp='200') # a dict of all trials for 1st net on cifar100, where the key is the seed
    #print ('There are {:} trials for this architecture [{:}] on cifar100'.format(len(results), api[1]))
    #print(results.keys())
    #print(api[i])
    v = []
    for seed, result in results.items():
        #print(result)
        #print(f'seed{seed}')
        #print ('Valid Info : {:}'.format(result.get_eval('x-valid')))
        v.append(result.get_eval('ori-test')['accuracy'])
    accuracy[api[i]]=(v)
```

```python
_opname_to_index = {
    'none': 0,
    'skip_connect': 1,
    'nor_conv_1x1': 2,
    'nor_conv_3x3': 3,
    'avg_pool_3x3': 4,
    'input': 5,
    'output': 6,
    'global': 7
}

_opindex_to_name = { value: key for key, value in _opname_to_index.items() }
```

```python
def get_arch_vector_from_arch_str(arch_str):
    ''' Args:
            arch_str : a string representation of a cell architecture,
                for example '|nor_conv_3x3~0|+|nor_conv_3x3~0|avg_pool_3x3~1|+|skip_connect~0|nor_conv_3x3~1|skip_connect~2|'
    '''

    nodes = arch_str.split('+')
    nodes = [node[1:-1].split('|') for node in nodes]
    nodes = [[op_and_input.split('~')[0]  for op_and_input in node] for node in nodes]

    # arch_vector is equivalent to a decision vector produced by autocaml when using Nasbench201 backend
    arch_vector = [_opname_to_index[op] for node in nodes for op in node]
    return arch_vector


def get_arch_str_from_arch_vector(arch_vector):
    ops = [_opindex_to_name[opindex] for opindex in arch_vector]
    return '|{}~0|+|{}~0|{}~1|+|{}~0|{}~1|{}~2|'.format(*ops)
```

```python
#translate to arch_vector
acc_d = {}
for k,v in accuracy.items():
    acc_d[tuple(get_arch_vector_from_arch_str(k))] = v
    
import pickle
pickle.dump(acc_d, open('nb2_test_acc_cf10.p','wb'))
```

```python
results = {}
for edge in range(6):
    for opidx in range(5):
        for k,v in acc_d.items():
            if k[edge] == opidx:
                if (edge,opidx) not in results:
                    results[(edge,opidx)] = []
                results[(edge,opidx)].extend(v)
```

```python
for k,v in results.items():
    print(f'edge = {k[0]} op = {_opindex_to_name[k[1]]} avg_accuracy = {round(stat.mean(v),2)}% stdev={round(stat.stdev(v),2)}%')
```

```python
best_ops = []
for edge in range(6):
    bop = -1
    bacc= -1
    for opidx in range(5):
        k=(edge,opidx)
        acc = stat.mean(results[k])
        if acc > bacc:
            bop = opidx
            bacc = acc
    best_ops.append(bop)
```
